```python
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
#loading data sets
column_names=["MPG","Cylinder","Displacement","Horsepower","Weight","Acceleration","Model Year","Origin"]
raw_dataset=pd.read_csv("auto-mpg.csv",names=column_names,
                        na_values="?",comment="\t", index_col=False,skiprows=1)
dataset=raw_dataset.copy()
```

```python
dataset.tail()
```

```python
dataset.isna().sum()
```

```python
dataset=dataset.dropna()
```

```python
dataset.isna().sum()
```

```python
origin=dataset.pop("Origin")
dataset['USA']=(origin==1)*1.0
dataset['Europe']=(origin==2)*1.0
dataset['Japan']=(origin==3)*1.0
dataset.tail()
```

```python
dataset.tail()
```

```python
dataset
```

```python
#Train Test Splitting
```

```python
#80% data-traing,20%-testing
train_dataset=dataset.sample(frac=0.8,random_state=0)
```

```python
#20% remaining in testing
test_dataset=dataset.drop(train_dataset.index)
```

```python
train_labels=train_dataset.pop('MPG')
test_labels=test_dataset.pop('MPG')
#seprating labels from the features
#MPG-miles per gallon
```

```python
train_stats=train_dataset.describe()
train_stats
```

```python
#normalization-feature scaling(scale down)
def norm(x):
    return(x-train_stats['mean'])/train_stats['std']

normed_train_data=norm(train_dataset)
normed_test_data=norm(test_dataset)
```

```python
#predictive model
def build_model():
    model=keras.Sequential([
        #layers of neural networks 64, first hidden layer
        #input=>len(no of col in dataset)
        #activation function which we aplly on each neuron
        #val for each neuron
        layers.Dense(64,activation="relu",input_shape=[len(train_dataset.keys())]),
        layers.Dense(64,activation="relu"),
        layers.Dense(1)
    ])
    optimizer=tf.keras.optimizers.RMSprop(0.001)
#loss function is used to calculated diffrence between actual and predictive value
    model.compile(loss='mse',
             #update weight of neural networks
             optimizer=optimizer,
             #mean absolute error
             #mean square error
             metrics=['mae','mse'])
    return model
```

```python
model=build_model()
model.summary()
```

```python
epochs=1000
history=model.fit(normed_train_data,train_labels,epochs=epochs)
```

```python
#model testing

```

```python
loss,mae,mse=model.evaluate(normed_test_data,test_labels)
```

```python
test_predictions=model.predict(normed_test_data).flatten()
test_predictions
```
